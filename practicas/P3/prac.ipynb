```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_hist(df, bins=30, figsize=(8, 4)):
    """
    Grafica histogramas con KDE para todas las columnas numéricas de un DataFrame.

    """
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    
    for col in num_cols:
        plt.figure(figsize=figsize)
        sns.histplot(df[col], kde=True, bins=bins, stat='density', color='skyblue', alpha=0.6)
        plt.title(f'Histograma y KDE de {col}')
        plt.xlabel(col)
        plt.ylabel('Densidad')
        plt.show()

# Ejemplo de uso:
plot_hist(df)

```

```python

import pandas as pd

def check_data_completeness_nombrecompleto(df:pd.DataFrame):
    """
     Descripcion:

        Da una descripcion de como esta estrucuturado nuestro df, conteo de nulos, datos estadisticos, y clasificacion de las columnas 
        en columnas discretas o continuas
    Args:
        df: Dataframe para analizar
    
    Retorna:
        resumen: Datafrme con el analisis 
    """
    resumen = pd.DataFrame()
    

    ## Procesamiento de Datos 
    resumen['nulos'] = df.isna().sum()
    resumen['completitud_%'] = (1 - df.isna().sum() / len(df)) * 100
    resumen['dtype'] = df.dtypes
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    resumen_dispersion = df[num_cols].describe().T[['min', 'max', 'mean', 'std']]
    resumen = resumen.join(resumen_dispersion, how='left')
    
    ## Clasificacion de Columnas
    resumen['tipo_columna'] = [
        'Continua' if (
            pd.api.types.is_numeric_dtype(df[col]) 
            and df[col].nunique() > 10
        ) else 'Discreta'
        for col in df.columns
    ]
    
    return resumen


```

```python
import pandas as pd

def calcular_IQR(df, columna, factor=1.5):
    """
    Calcula el IQR y detecta outliers para una columna numérica.
    
    Parámetros:
    df : DataFrame
        DataFrame que contiene la columna.
    columna : str
        Nombre de la columna numérica.
    factor : float, opcional (default=1.5)
        Factor para determinar los límites de los outliers.
        
    Returns:
    dict con:
        - Q1, Q3, IQR
        - límite inferior y superior
        - outliers
    """
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    
    limite_inferior = Q1 - factor * IQR
    limite_superior = Q3 + factor * IQR
    
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)][columna]
    
    return {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'limite_inferior': limite_inferior,
        'limite_superior': limite_superior,
        'outliers': outliers
    }

# Ejemplo de uso:
resultado = calcular_IQR(df, 'columna_numerica')
print(resultado['limite_inferior'], resultado['limite_superior'])
print("Outliers encontrados:\n", resultado['outliers'])
```
